```python
#Gerekli kutuphaneleri yukluyoruz.
import matplotlib.pyplot as plt   #goruntuleri cizdirmek icin
import numpy as np                #sayisal islemler icin
import cv2
from google.colab import files    #dosya yuklemek icin
```

```python
#Fotograflari okuma ve RGB formatina cevirme
day_images=['Grup10_resim1d.jpeg','Grup10_resim1m.jpeg','Grup10_resim1u.jpeg']
night_images=['Grup10_resim2d.jpeg','Grup10_resim2m.jpeg','Grup10_resim2u.jpeg']

day_photos = []
for img_name in day_images:
    img = cv2.imread(img_name)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    day_photos.append(img_rgb)


night_photos = []
for img_name in night_images:
    img = cv2.imread(img_name)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    night_photos.append(img_rgb)

print("Toplam", len(day_photos), "gündüz fotoğrafı ve", len(night_photos), "gece fotoğrafı yüklendi.")
```

```python
#Gunduz fotograflarinin her birinden 24x24'luk kucuk bolge secip gosterelim
plt.figure(figsize=(15, 5))
for i, img in enumerate(day_photos):
  crop = img[0:24, 0:24]

# Alt alta 1 satir 3 sutun olacak sekilde subplot yapiyoruz
plt.subplot(1, 3, i + 1)
plt.imshow(crop)
plt.title(f"Gündüz Fotoğrafı {i + 1}")
plt.axis('off')

plt.show()
```

```python
#Tum gunduz fotograflarinin renk kanallarini ayirma ve gosterme
plt.figure(figsize=(15, 15))

for i, img in enumerate(day_photos):
    # Renk kanallarini ayiralim
    red_channel = img[:, :, 0]
    green_channel = img[:, :, 1]
    blue_channel = img[:, :, 2]

    # Her fotograf icin renk kanallarini yan yana gosterelim
    plt.subplot(len(day_photos), 3, i * 3 + 1)
    plt.imshow(red_channel, cmap='Reds')
    plt.title(f"Fotoğraf {i + 1} - Kırmızı Kanal")
    plt.axis('off')

    plt.subplot(len(day_photos), 3, i * 3 + 2)
    plt.imshow(green_channel, cmap='Greens')
    plt.title(f"Fotoğraf {i + 1} - Yeşil Kanal")
    plt.axis('off')

    plt.subplot(len(day_photos), 3, i * 3 + 3)
    plt.imshow(blue_channel, cmap='Blues')
    plt.title(f"Fotoğraf {i + 1} - Mavi Kanal")
    plt.axis('off')

plt.show()
```

```python
#Piksel degerlerini yazdirma islemi
#24x24'luk bir bolge
crop_size = 24

# Gunduz fotograflarinin her biri için renk kanalındaki pikselleri yazdiralim
for i, img in enumerate(day_photos):

    red_channel = img[:, :, 0]
    green_channel = img[:, :, 1]
    blue_channel = img[:, :, 2]

    # 24x24'luk bir bolgeyi secelim
    crop_red = red_channel[0:crop_size, 0:crop_size]
    crop_green = green_channel[0:crop_size, 0:crop_size]
    crop_blue = blue_channel[0:crop_size, 0:crop_size]

    print(f"Fotoğraf {i + 1} - Kırmızı Kanal (Pikseller):")
    print(crop_red)
    print(f"Fotoğraf {i + 1} - Yeşil Kanal (Pikseller):")
    print(crop_green)
    print(f"Fotoğraf {i + 1} - Mavi Kanal (Pikseller):")
    print(crop_blue)
    print("\n" + "="*40 + "\n")
```

```python
#Resimleri farklı renk haritasi ile(color map) kullanarak gosterme
plt.figure(figsize=(15, 15))

for i, img in enumerate(day_photos):

    red_channel = img[:, :, 0]
    green_channel = img[:, :, 1]
    blue_channel = img[:, :, 2]

    # Her fotograf için renk kanallarini farklı renk haritalariyla gosterelim
    plt.subplot(len(day_photos), 3, i * 3 + 1)
    plt.imshow(red_channel, cmap='hot')  # Kırmızı kanal için 'hot' renk haritası
    plt.title(f"Fotoğraf {i + 1} - Kırmızı Kanal")
    plt.axis('off')

    plt.subplot(len(day_photos), 3, i * 3 + 2)
    plt.imshow(green_channel, cmap='winter')  # Yeşil kanal için 'winter' renk haritası
    plt.title(f"Fotoğraf {i + 1} - Yeşil Kanal")
    plt.axis('off')

    plt.subplot(len(day_photos), 3, i * 3 + 3)
    plt.imshow(blue_channel, cmap='cool')  # Mavi kanal için 'cool' renk haritası
    plt.title(f"Fotoğraf {i + 1} - Mavi Kanal")
    plt.axis('off')

plt.show()
```

```python
#Gece fotograflarini yan yana goruntuleme ve her biri icin histogram olusturma
# Gece fotograflarini yan yana gosterelim
plt.figure(figsize=(15, 5))
for i, img in enumerate(night_photos):
    plt.subplot(1, 3, i + 1)
    plt.imshow(img)
    plt.title(f"Gece Fotoğrafı {i + 1}")
    plt.axis('off')
plt.show()

#Her gece fotografi icin histogramlari olusturup yan yana cizdiriyoruz
plt.figure(figsize=(15, 5))
for i, img in enumerate(night_photos):
    plt.subplot(1, 3, i + 1)
    # img.ravel() ile tüm pikselleri tek boyutlu hale getirip histogramını oluşturuyoruz
    plt.hist(img.ravel(), bins=256, color='gray', alpha=0.7)
    plt.title(f"Gece Fotoğrafı {i + 1} Histogram")
    plt.xlabel("Piksel Değeri")
    plt.ylabel("Frekans")
plt.show()
```

Histogram grafiği,dağılımın yoğunluğuna ve dağılımın yaygınlığına bakarak yorumlanabilir.

Dağılım yoğunluğu,histogramın belirli bir bölgeye (örneğin 0-50 arası koyu tonlar) yoğunlaşması, görüntünün genel olarak karanlık olduğunu bilgisini bize verir.

Dağılım yaygınlığı ise histogram tüm değer aralığını (0-255) kapsıyorsa, görüntüde geniş bir kontrast aralığı var bilgisini verir.

Her üç gece fotoğrafı için elde edilen histogramları yorumlayalım.

1.Gece Fotoğrafı

Histogramda piksel değerlerinin büyük çoğunluğu düşük parlaklık seviyesinde yoğunlaşmış. Görüntünün genel olarak karanlık olduğunu gösterir.

2.Gece Fotoğrafı

Histogramda yine düşük parlaklık seviyelerine odaklanmış ancak biraz daha geniş bir dağılım olduğunu görüyoruz.

3.Gece Fotoğrafı

En keskin histogramlardan biri, tamamen düşük parlaklık seviyelerine yoğunlaşmış.


Fotoğrafları parlaklık-kontrast ayarlarıyla, histogram eşitleme, gürültü azaltma ve keskinlik arttırarak iyileştirebiliriz.
